```python
#importing all the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset

df=pd.read_csv("Zomato.csv")
```

```python
df.columns
```

# Data cleaning

```python
#removing the unnecessary columns
df=df.drop(['url','address','phone','menu_item','listed_in(type)','listed_in(city)'],axis=1)
```

```python
df.shape
```

```python
df.isnull().sum()
```

Here we can see there are many null values present in the columns like 'rate','location','rest_type','dish_liked','cuisines'and 'approx_cost(for two people)'

```python
#dropping the null values from the rate column
df.dropna(subset=['rate'],axis=0,inplace=True)
```

```python
def split(x):
    return x.split('/')[0].strip()
```

```python
df['rate'] = df['rate'].apply(split)
df['rate']
```

```python
df['rate'].unique()
```

```python
#replacing the 'NEW','-' from the rate column with '0'

df['rate'].replace(['NEW','-'],0,inplace=True)
df['rate']=df['rate'].astype(float)
df['rate']
```

```python
df.info()
```

# exploratory data analysis

```python
#calculating the average rating for top 15 restaurants with highest average rating
rating = pd.pivot_table(df,index='name',values='rate')
rating=rating.sort_values(['rate'],ascending=False).head(15)
rating
```

```python
#barplot of top 15 restaurants with highest average ratings

plt.figure(figsize=(12,6))
sns.barplot(x=rating.rate,y=rating.index,orient="h")
plt.show()
```

from the above bar plot we can see that Asia kitchen by mainland china and Byg Brewski Brewing company have the highest average rating. 

```python
sns.set_style('whitegrid')
sns.distplot(df['rate'])

plt.show()
```

from the above graph we can clearly see that the rate is not normally distributed.

```python
#Analysing restaurant chains
def produce_chains(name):
    data_chain=pd.DataFrame(df[df["name"]==name].value_counts().reset_index())
    data_chain.columns=['Name','count']
    return data_chain[['Name','count']]


df_1=df.groupby(['rest_type','name']).agg('count')
chain_data=df_1.sort_values(['cuisines'],ascending=False).groupby(['rest_type'],
                                                        as_index=False).apply(lambda x : x.sort_values(by="cuisines",ascending=False).head(15))['cuisines'].reset_index().rename(columns={'cuisines':'count'})
```

```python
#analyzing top 15 casual dining restaurant chsains
casual=chain_data[chain_data['rest_type']=='Casual Dining']
casual
```

```python
#Top 15 casual dining restaurant chains in Banglore
plt.figure(figsize=(12,6))
sns.barplot(x=casual['name'],y=casual['count'],palette='magma')
plt.xlabel('name of chains')
plt.ylabel('Number of Restaurants')
plt.title("most popular casual dining restaurant chains")
plt.xticks(rotation=90)
```

Empire Restaurant has the most number of outlets among casual dining restaurants chains.

```python
#analyzing top 15 quick bites restaurants
quick_bites=chain_data[chain_data['rest_type']=='Quick Bites']
quick_bites
```

```python
#Top 15 quick_bites restaurant chains in Banglore
plt.figure(figsize=(12,6))
sns.barplot(x=quick_bites['name'],y=quick_bites['count'],palette='magma')
plt.xlabel('name of chains')
plt.ylabel('Number of Restaurants')
plt.title("most popular casual dining restaurant chains")
plt.xticks(rotation=90)
```

# Analyzing the cost for two people

```python
df['approx_cost(for two people)'].isnull().sum()
```

```python
df.dropna(subset=['approx_cost(for two people)'],axis=0,inplace=True)
```

```python
df['approx_cost(for two people)']
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(lambda x: x.replace(',',''))
```

```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype(int)
```

```python
df['approx_cost(for two people)'].describe()
```

```python
sns.distplot(df['approx_cost(for two people)'])
plt.show()
```

from the distribution plot we can say that the cost for two people in most of restaurants is below 1000 Rupees
